# Final Data Preparation for Machine Learning(Data Cleaning)

```python
#to check how many null values are there
data_train.isnull().sum()
```

```python
data_train[data_train.Age.isnull()]
```

```python
#removing the name and only keeping the titles
test_train=[data_train,data_test]
for dataset in test_train:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
data_train['Title'].value_counts()
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in test_train:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
data_train.head()
#from graph,3 and 2 are more likely to survive
```

```python
# delete unnecessary feature from dataset# delete 
data_train.drop('Name', axis=1, inplace=True)
data_test.drop('Name', axis=1, inplace=True)
```

```python
data_train.head()
```

```python
sex_mapping = {"male": 0, "female": 1}
for dataset in test_train:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
```

```python
data_train.head()
#1 is more likely to survive
```

```python
# fill missing age with median age for each title (Mr, Mrs, Miss, Others)
data_train["Age"].fillna(data_train.groupby("Title")["Age"].transform("median"), inplace=True)
data_test["Age"].fillna(data_test.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
data_train.head(30)
data_train.groupby("Title")["Age"].transform("median")
```

```python
# fill missing age with median age for each title (Mr, Mrs, Miss, Others)# fill m 
data_train["Age"].fillna(data_train.groupby("Title")["Age"].transform("median"), inplace=True)
data_test["Age"].fillna(data_test.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
#data binning feature vector map:
"""child: 0
young: 1
adult: 2
mid-age: 3 senior: 4"""
for dataset in test_train:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0,
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1,
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2,
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3,
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
```

```python
#since most of them are from S embark,filling all missing embarked with S
for dataset in test_train:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
```

```python
#marking embark
embarked_mapping = {"S": 0, "C": 1, "Q": 2}
for dataset in test_train:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_mapping)
```

```python
# fill missing Fare with median fare for each Pclass
data_train["Fare"].fillna(data_train.groupby("Pclass")["Fare"].transform("median"), inplace=True)
data_test["Fare"].fillna(data_test.groupby("Pclass")["Fare"].transform("median"), inplace=True)
```

```python
#data binning for fare
for dataset in test_train:
    dataset.loc[ dataset['Fare'] <= 17, 'Fare'] = 0,
    dataset.loc[(dataset['Fare'] > 17) & (dataset['Fare'] <= 30), 'Fare'] = 1,
    dataset.loc[(dataset['Fare'] > 30) & (dataset['Fare'] <= 100), 'Fare'] = 2,
    dataset.loc[ dataset['Fare'] > 100, 'Fare'] = 3
```

```python
data_train.head()
```

```python
for dataset in test_train:
    dataset['Cabin'] = dataset['Cabin'].str[:1]
```

```python
cabin_mapping = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for dataset in test_train:
    dataset['Cabin'] = dataset['Cabin'].map(cabin_mapping)
```

```python
data_train['Cabin'].head(10)
```

```python
# fill missing cabin with median cabin for each Pclass 
data_train["Cabin"].fillna(data_train.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
data_test["Cabin"].fillna(data_test.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
```

```python
data_train.head()
```

```python
data_train["FamilySize"] = data_train["SibSp"] + data_train["Parch"] + 1
data_test["FamilySize"] = data_test["SibSp"] + data_test["Parch"] + 1
```

```python
family_mapping = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}
for dataset in test_train:
    dataset['FamilySize'] = dataset['FamilySize'].map(family_mapping)
```

```python
#removing SibSp,Parch and Passengerid:
data_train=data_train.drop(['PassengerId','SibSp','Parch'],axis=1)
data_test=data_test.drop(['SibSp','Parch'],axis=1)
```

```python
data_test=data_test.drop(['Ticket'],axis=1)
data_train=data_train.drop(['Ticket'],axis=1)
data_train.head()
```

```python
data_test.head()
```

# Modelling

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

import numpy as np
```

```python
data_train.info()
```

```python
traindata = data_train.drop('Survived', axis=1)
target = data_train['Survived']

traindata.shape, target.shape
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
#Random forest classifier
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, traindata, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score)*100, 2)
```

```python
#SVM:Support vector mechanism
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, traindata, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```
